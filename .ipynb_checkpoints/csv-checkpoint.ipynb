```python
# list inside dicts
# {'Name':[],'Age':[],'Add':[]}
import pandas as pd
infos = {}

name = []
age = []
add = []
n = int(input("Enter n = "))
for i in range(n):
    na = input("ENter name = ")
    ag = input("ENter age = ")
    ad = input("Enter address = ")
    
    name.append(na)
    age.append(ag)
    add.append(ad)
    
    
infos['Name']=name
infos['Age'] = age
infos['Address']=add
    
print(infos)


df = pd.DataFrame(infos)
df.to_csv('NewFile.csv')
```
